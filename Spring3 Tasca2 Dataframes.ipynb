```python
f_clean.describe()
```

```python
f_clean.isna().sum().sort_values(ascending=False)
```

```python
#Average speed

milestokmh= 60/0.62137

f_clean['Average_Speed']=(f_clean['Distance'] / f_clean['ActualElapsedTime'])*milestokmh
f_clean['Average_Speed'].round(2).head(3)
```

```python
# Delayed

f_clean['Delayed']=f_clean['ArrDelay']>0
f_clean['Delayed'].sort_values(ascending=False)
```

```python
f_clean['Delayed'] = np.where(f_clean['ArrDelay'] > 0, 1, 0)
f_clean['Delayed'].sum()
```

```python
# Acum Delayed

acumDelayed=f_clean.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False)
acumDelayed.head(10)
```

```python
# Longer flights

Longer=f_clean[['ActualElapsedTime','UniqueCarrier','FlightNum','TailNum', 'Distance']].copy()
Longer.sort_values(by='ActualElapsedTime', ascending=False).head()
```

```python

delayed_flights=f_clean[['ArrDelay','UniqueCarrier','FlightNum','TailNum', 'Distance']].copy()
delayed_flights.sort_values(by='ArrDelay', ascending=False).head(7)
```
